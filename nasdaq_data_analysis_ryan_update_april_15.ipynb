```python
#Creation of variable for 2019 Data.
ndx_2019 = ndx_df[(ndx_df["Date"] >= "2019-01-01") & (ndx_df["Date"] <= "2019-12-31")]

ndx_2019.head()
```

```python
#Correlation Plot of 2019 Date and Volume
plt.scatter(ndx_2019["Date"], ndx_2019["Volume"], color="Blue")
plt.title("2019 Date to Volume Correlation")
plt.xlabel("Date")
plt.ylabel("Volume (Billions)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

#Correlation Practice, doesn't really tell you anything.
ndx_2019_correlation_construct = ["Date", "Open", "Close", "High", "Low", "Adj Close", "Volume"]
ndx_2019_correlation_calculation = ndx_2019[ndx_2019_correlation_construct].corr()
display(ndx_2019_correlation_calculation)
```
